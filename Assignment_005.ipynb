4. What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

```python
gender_score_variation = data1.groupby('gender').agg({'math score': ['std', 'var', lambda x: x.max() - x.min()],
                                                   'reading score': ['std', 'var', lambda x: x.max() - x.min()],
                                                   'writing score': ['std', 'var', lambda x: x.max() - x.min()]})
gender_score_variation
```

```python
test_variation = data1.groupby('test preparation course').agg({'math score': ['std', 'var', lambda x: x.max() - x.min()],
                                                   'reading score': ['std', 'var', lambda x: x.max() - x.min()],
                                                   'writing score': ['std', 'var', lambda x: x.max() - x.min()]})
test_variation
```

5. The management needs your help to give bonus points to the top 25% of
students based on their math score, so how will you help the management
to achieve this.


```python
threshold_score = data1['math score'].quantile(0.75)
qualifying_students = data1[data1['math score'] >= threshold_score]
bonus_points = 5
qualifying_students['bonus_points'] = bonus_points
qualifying_students
```
